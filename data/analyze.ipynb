```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('question_data.csv')
df
```

```python
data = df.groupby(['chapter', 'level'])['code'].count().reset_index()
data
```

```python
diff = 0
for i in data.values:
    if i[0] != diff:
        print('='*100)
    print('Chapter {0[0]}, level {0[1]}: {0[2]}'.format(i))
    diff = i[0]
```

```python
fig, ax = plt.subplots(figsize=(10, 20))
sns.barplot(data=data, y='chapter', x='code', hue='level', ax=ax)
plt.show()
```
